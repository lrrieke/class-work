```python
import pandas as pd
%matplotlib inline 
```

```python
df = pd.read_csv('8-1_Multi.csv', thousands=',')
df
```

```python
df.plot()
```

```python
df.plot(kind = 'bar', title = "2017 Sales");
```

```python
p = df.plot(kind = 'bar');
p.set_title('2017 Sales', fontsize=24, color = 'Purple');
p.set_ylabel('Units Sold', fontsize=12, color = 'blue');
p.set_xlabel('Month', fontsize=12, color = 'green');
```

```python
p = df.plot(kind = 'bar', stacked=True);
p.set_title('2017 Sales', fontsize=24);
p.set_ylabel('Units Sold', fontsize=12);
p.set_xlabel('Month', fontsize=12);
```
