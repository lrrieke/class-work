```python
import pandas as pd

ufo = pd.read_csv("uforeports.csv")

ufo.head()
```

When were the earliest and latest reported sitings?

```python
# convert 'Time' to datetime format

ufo['Time'] = pd.to_datetime(ufo.Time)
```

```python
# run ufo head to make sure it converted

ufo.head()
```

```python
# use max() to get latest siting

ufo.Time.max()
```

```python
# looks like the most recent reported siting for this dataset was New Year's Eve in 2000
# now I'll move on to earliest reported siting (even you can see it in the head)

ufo.Time.min()
```

Which date has the most reported sitings?

```python
ufo['Time'].describe()
```

What color is most frequently seen in reported sitings?

```python
# I'll define the column I'm looking at and run a frequency check using the describe() function

ufo['Colors Reported'].describe()
```

Which city has the most reported sitings?

```python
# same function as above, just a different argument supplied

ufo['City'].describe()
```

Okay, enough of the describe function. I wonder if I can group sitings by city instead of date.

```python
# use the groupby function

g = ufo.groupby('City')
g
```

```python
# looks like that worked, let's double check

for City, City_df in g:
    print(City)
    print(City_df)
```

This is an absolutely huge dataframe, but grouping by city did work! Now I wonder if I can just see New York's sitings.

```python
#I decided to ask some specific questions about New York, so I stored the group as a new dataframe

ufo2 = g.get_group('New York City')

ufo2
```

What shape is most frequently reported in New York's sitings?

```python
#I'll use the same describe() function as above

ufo2['Shape Reported'].describe()
```
